```python
from statistics import mode 
def get_consensus(result_arr, top_k=[ None ]):
    embeddings, profiles, distances = list( zip(*result_arr) )
    
    ret = ()
    for k in top_k:
        ret = ret + ( mode(profiles[0:k]), )
    return ret
```

```python
for key, value in import_result.items():
    predictions = get_consensus( value,  list(range(1, 21)) )
    df.at[key, 'prediction'] = True
    for n in range(1, 21):
        df.at[key, f'predict_k_{n}'] = predictions[n - 1]
```
